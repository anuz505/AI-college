```python
import matplotlib.pyplot as plt
from scipy import stats
```

```python
x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,100,86,103,87,94,78,77,85,86]
plt.scatter(x,y)
plt.show()
```

```python
slope,intercept,r,p,std_err = stats.linregress(x,y) 
# returns the slope,intercept,rvalue,pvalue,stderr # do check this 
# rvalue is the correlation coefficient
# pvalue is the probability value
# stderr is the standard error



def myfunc(x):
    return slope * x + intercept

my_model = list(map(myfunc,x))
print(my_model)
plt.scatter(x,y)
plt.plot(x,my_model)
plt.show()
```

# Linear Regression Summary

- **Slope (m):** -1.4558498896247238
  
- **Intercept (b):** 100.00993377483444


- **Correlation Coefficient (r):** -0.7641  
- **P-value:** 0.0024  
- **Standard Error:** 0.37054947000125027


### Interpretation
- The correlation coefficient (**r**) indicates a **strong negative linear relationship** between `x` and `y`.  
- The **p-value** (< 0.05) suggests the relationship is **statistically significant**, meaning the slope is not zero.  
- The **standard error** is a measure of the accuracy of the coefficient estimates.

The standard error (SE) measures the accuracy of the slope estimate in a linear regression. It tells you how much the estimated slope would vary if you repeated the analysis with different samples.


A small SE (e.g., close to 0) means the slope estimate is precise.
A larger SE indicates more uncertainty in the slope estimate.
If your standard error = 0.3705, it means there is moderate variability in the slope estimate, but its significance should also be assessed using the p-value (which is already small, so your slope is likely meaningful).
